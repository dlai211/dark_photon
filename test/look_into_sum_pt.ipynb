```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import awkward as ak
```

```python
path = "/data/tmathew/ntups/mc23d/ggHyyd_y.root"
f = uproot.open(path)['nominal']
f.keys()

vtx_sumPt = f['vtx_sumPt'].array(library='ak')
print(ak.flatten(vtx_sumPt))

plt.figure(figsize=(20, 7))
plt.subplot(121)
plt.hist(ak.flatten(vtx_sumPt), histtype='step', bins=100)
plt.xlabel('vtx_sumPt (GeV)')
plt.grid()


plt.subplot(122)
plt.hist(ak.flatten(vtx_sumPt), histtype='step', bins=np.linspace(0, 100, 100))
plt.title('Zoom in [0, 100]')
plt.xlabel('vtx_sumPt (GeV)')

plt.grid()
plt.show()
```
